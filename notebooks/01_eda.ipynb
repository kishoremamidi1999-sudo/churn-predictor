```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Makes charts look better
sns.set_theme(style="whitegrid")
print("Libraries loaded!")
```

```python
#2a. Load the data
df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
print(f"Shape: {df.shape}")  # rows x columns
df.head()
```

```python
# 2b. Fix Column Names
df.columns = df.columns.str.strip()
print("Columns after fix:")
print(df.columns.tolist())
```

```python
# 3. Understand the columns
df.info()
```

```python
# 4. Check for missing values
missing = df.isnull().sum()
print(missing[missing > 0]) #only show columns that have missing values
```

```python
# 5. Check the target column (Churn)
print(df["Churn"].value_counts())
print("\nChurn rate:", df["Churn"].value_counts(normalize=True).round(3))
print("\nUnique values:", df["Churn"].unique())  # add this to see exact values
```

```python
# 6. visualize churn distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="Churn", data=df, palette="Set2")
plt.title("Churn Distribution")
plt.xlabel("Churn (Yes = Left, No = Stayed)")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.savefig("../data/churn_distribution.png")  # saves the chart
plt.show()
```

```python
# 7. churn by contract type
plt.figure(figsize=(8, 5))
sns.countplot(x="Contract", hue="Churn", data=df, palette="Set2")
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("../data/churn_by_contract.png")
plt.show()
```

```python
# cell8. Monthly Charges vs Churn
plt.figure(figsize=(8, 5))
sns.boxplot(x="Churn", y="MonthlyCharges", data=df, palette="Set2")
plt.title("Monthly Charges Vs Churn")
plt.tight_layout()
plt.savefig("../data/charges_vs_churn.png")
plt.show()
```

```python
# Cell9. Correlation heatmap(numeric columns)
numeric_df = df.select_dtypes(include=["float64", "int64"])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt="2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("../data/correlation_heatmap.png")
plt.show()
```
