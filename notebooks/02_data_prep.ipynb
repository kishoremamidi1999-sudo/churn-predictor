```python
# cell1. Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import pickle
df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.columns = df.columns.str.strip() #fix column name spaces
print("Shape:", df.shape)
df.head()
```

```python
# Cell2. Drop customerID
df.drop(columns=["customerID"], inplace=True)
print("customerID column removed")
df.head()
```

```python
# cell 3. Fix TotalCharges(it's stored as text, Should be number)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
print("Missing values after fix:", df["TotalCharges"].isnull().sum())

```

```python
# Cell 4. Fill missing TotalCharges with median
df["TotalCharges"].fillna(df["TotalCharges"].median(), inplace=True)
print("Missing values now:", df["TotalCharges"].isnull().sum())
```

```python
# Cell 5. convert Target Column
# convert churn Yes/No to 1/0
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
print(df["Churn"].value_counts())
```

```python
# Cell 6. Encode Categorical Columns
le = LabelEncoder()
categorical_cols = df.select_dtypes(include=["object"]).columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
print("Done encoding!")
df.head()
```

```python
# Cell 7. Separate input features (X) and target (y)
X = df.drop(columns=["Churn"])
Y = df["Churn"]
print("X shape:", X.shape)
print("Y shape:", Y.shape)
```

```python
print(df["Churn"].isna().sum())
```

```python
# Cell 8. Split into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
print("Training rows:", X_train.shape[0])
print("Testing rows:", X_test.shape[0])
```
